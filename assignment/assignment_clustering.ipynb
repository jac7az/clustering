**Q1.** This is a question about clustering. We want to investigate how adjusting the "noisiness" of the data impacts the quality of the algorithm and the difficulty of picking $k$.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

```
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)

df0_125 = createData(0.125)

```

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?
3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
4. Explain the intuition of the elbow, using this numerical simulation as an example.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
df0_125
````

````python
sns.scatterplot(df0_125,x='x1',y='x2',hue='group')
plt.title('df0_125')
plt.show()

sns.scatterplot(df0_25,x='x1',y='x2',hue='group')
plt.title('df0_25')
plt.show()

sns.scatterplot(df0_5,x='x1',y='x2',hue='group')
plt.title('df0_5')
plt.show()

sns.scatterplot(df1_0,x='x1',y='x2',hue='group')
plt.title('df1_0')
plt.show()

sns.scatterplot(df2_0,x='x1',y='x2',hue='group')
plt.title('df2_0')
plt.show()

#As noise level increases, the visual distinction between each of the 3 groups becomes more mixed and indistinguishable
````

````python
def maxmin(x): # Max-min scaler function
    x = (x-min(x))/(max(x)-min(x))
    return x
df0_125=df0_125[['x1','x2']].apply(maxmin)
df0_25=df0_25[['x1','x2']].apply(maxmin)
df0_5=df0_5[['x1','x2']].apply(maxmin)
df1_0=df1_0[['x1','x2']].apply(maxmin)
df2_0=df2_0[['x1','x2']].apply(maxmin)
SSE125=[]
SSE25=[]
SSE5=[]
SSE1=[]
SSE2=[]
for k in range(10):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(df0_125)
    SSE125.append(model.inertia_)
    model = model.fit(df0_25)
    SSE25.append(model.inertia_)
    model = model.fit(df0_5)
    SSE5.append(model.inertia_)
    model = model.fit(df1_0)
    SSE1.append(model.inertia_)
    model = model.fit(df2_0)
    SSE2.append(model.inertia_)

sns.lineplot(x=np.arange(1,11), y=SSE125,label='df0_125').set_title('Scree Plot')
plt.plot(np.arange(1,11),SSE25,label='df0_25')
plt.plot(np.arange(1,11),SSE5,label='df0_5')
plt.plot(np.arange(1,11),SSE1,label='df1_0')
plt.plot(np.arange(1,11),SSE2,label='df2_0')
plt.legend()
plt.xlabel('Clusters')
plt.ylabel('SSE Values')
plt.show()

#The "elbow" of this plot represents the optimal cluster, where SSE drops to the lowest without being to overfit, in which it would only be applicable to the training dataset.
#For df1_0 and df2_0, optimal cluster number is 2. For the other 3, df0_125, df0_25 and df0_5, optimal cluster appears to be at 3.
````

````python
url='https://raw.githubusercontent.com/jac7az/clustering/refs/heads/main/assignment/data/2022%20election%20cycle%20fundraising.csv'
election=pd.read_csv(url)

#Isolate the 4 columns first into its own dataset then, take out all $ signs and commas, then convert to integers for all columns.
  #Finally, drop all rows with na. In this case, there are none.
election=election[['Raised','Spent','Cash on Hand','Debts']]
election['Raised']=pd.to_numeric(election['Raised'].replace({"\$":'',',':''},regex=True).dropna())
election['Spent']=pd.to_numeric(election['Spent'].replace({"\$":'',',':''},regex=True).dropna())
election['Cash on Hand']=pd.to_numeric(election['Cash on Hand'].replace({"\$":'',',':''},regex=True).dropna())
election['Debts']=pd.to_numeric(election['Debts'].replace({"\$":'',',':''},regex=True).dropna())

def maxmin(x):
    x = (x-min(x))/(max(x)-min(x))
    return x

election1=election[['Raised','Spent']].apply(maxmin)

SSE=[]
for k in range(10):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(election1)
    SSE.append(model.inertia_)

sns.lineplot(x=np.arange(1,11), y=SSE).set_title('Scree Plot')
plt.xlabel('Clusters')
plt.ylabel('SSE Values')
plt.show()

kmeans=KMeans(n_clusters=3,random_state=0).fit(election1[['Raised','Spent']])
election1['Cluster']=kmeans.labels_

sns.scatterplot(election1,x='Raised',y='Spent',hue='Cluster')
plt.title('Spent v. Raised')
plt.show()

#The overall relationship has a pretty strong linear correlation. As raised amount increases, the increment between the next spent value spreads out more
  #In the 3 clusters, raised value up to 0.2 is in a cluster, 0.2-0.5 are clustered together, and 0.5+ are in one cluster
````

````python
election2=election[['Cash on Hand','Debts']].apply(maxmin)

SSE_e=[]
for k in range(10):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(election2)
    SSE_e.append(model.inertia_)

sns.lineplot(x=np.arange(1,11), y=SSE_e).set_title('Scree Plot')
plt.xlabel('Clusters')
plt.ylabel('SSE Values')
plt.show()
kmeans=KMeans(n_clusters=3,random_state=0).fit(election2[['Cash on Hand','Debts']])
election2['Cluster']=kmeans.labels_

sns.scatterplot(election2,x='Cash on Hand',y='Debts',hue='Cluster')
plt.title('Cash on Hand v. Debts')
plt.show()

#For this plot, there's is a much much weaker correlation between debts and cash on hand compared to spent v. raised. The cluster range is still about 0-0.2, 0.2-0.5 & 0.5+ for Cash on Hand while the debt values vary much more.
  #Most people don't have debt or much debt.
````

````python
election=election[['Raised','Spent','Cash on Hand','Debts']].apply(maxmin)
SSE=[]
for k in range(10):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(election)
    SSE.append(model.inertia_)

sns.lineplot(x=np.arange(1,11), y=SSE).set_title('Scree Plot')
plt.xlabel('Clusters')
plt.ylabel('SSE Values')
plt.show() #k=2

kmeans = KMeans(n_clusters=4, random_state=0).fit(election)
election['Cluster']=kmeans.labels_
election

sns.scatterplot(election,x='Cash on Hand',y='Spent',hue='Cluster')
plt.title('Spent v. Cash on Hand')
plt.show()
sns.scatterplot(election,x='Cash on Hand',y='Raised',hue='Cluster')
plt.title('Raised v. Cash on Hand')
plt.show()
sns.scatterplot(election,x='Debts',y='Spent',hue='Cluster')
plt.title('Spent v. Debt')
plt.show()
sns.scatterplot(election,x='Debts',y='Raised',hue='Cluster')
plt.title('Raised v. Debt')
plt.show()

#The new best cluster is at k=4. The relation between cash on hand with raised and spent are incredibly similar, which makes sense since raised and spent have a strong linear correlation
  #Most people don't have any debt, so it's all cluster towards debts=0.
#All 4 clusters have multiple points in each group, but cluster 1 has the least, with only 3, which
  #may imply that this cluster doesn't provide much meaningful information.
````
